The case of the disappearing teaspoons: longitudinal cohort study of the displacement of teaspoons in an Australian research institute

```python
import numpy as np
import matplotlib.pyplot as plt
```

This is a binomial distribution
with
Independent trials: Each trial (whether a teaspoon disappears in a

*   Independent trials- each trial is of one spoon and whether it disappears
*   Fixed number of trials- # of trials per week is equal to the # of spoons present at the start of the week
*   Probability of success- each easpoon has the same probability of disappearing in a given week

Alternatively it may be an exponential distribution with a half life of 81 days as by 81 days half of the spoons were missing, but more data would have to be collected to see if that matched


```python
# Parameters
initial_teaspoons = 100  # Starting number of teaspoons
days = 147  # Duration of the study in days (20 weeks = 140 days)
disappearance_prob_weekly = 0.10  # Probability of a teaspoon disappearing each week

# Calculate daily disappearance probability
disappearance_prob_daily = 1 - (1 - disappearance_prob_weekly) ** (1 / 7)

# Simulate the number of teaspoons left each day
teaspoons_left = [initial_teaspoons]
for day in range(1, days + 1):
    teaspoons_left.append(teaspoons_left[-1] - np.random.binomial(teaspoons_left[-1], disappearance_prob_daily))

# Plot the results
plt.figure(figsize=(10,6))
plt.plot(range(days + 1), teaspoons_left, marker='o')
plt.title('Simulated Teaspoon Disappearance Over Time (Daily)')
plt.xlabel('Days')
plt.ylabel('Number of Teaspoons Left')
plt.grid(True)
plt.show()
```

Alternatively it may be an exponential distribution with a half life of 81 days as by 81 days half of the spoons were missing, but more data would have to be collected to see if that matched


```python

# Parameters
initial_teaspoons = 70  # Starting number of teaspoons
half_life = 81  # Half-life of teaspoons in days
days = 147  # Duration of the study in days

# Calculate the decay constant k based on the half-life
k = np.log(2) / half_life

# Simulate the number of teaspoons left over time using exponential decay
time = np.arange(0, days + 1)
teaspoons_left = initial_teaspoons * np.exp(-k * time)

# Plot the results
plt.figure(figsize=(10,6))
plt.plot(time, teaspoons_left, marker='o')
plt.title('Teaspoon Disappearance Over Time (Exponential Decay)')
plt.xlabel('Days')
plt.ylabel('Number of Teaspoons Left')
plt.grid(True)
plt.show()
```
